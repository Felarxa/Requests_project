```python
api_base_url = 'https://api.github.com'
owner = 'amzn'
url = f'{api_base_url}/users/{owner}/repos'
url
```

```python
repos_list = []
for page_num in range(1,6):
    try:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

```python
repos_list
```

```python
len(repos_list)
```

# Transform

```python
repos_list[0][0]
```

```python
repos_list[0][0].keys()
```

```python
repos_name = []
for page in repos_list:
    for repo in page:
        repos_name.append(repo['name'])
```

```python
len(repos_name)
```

```python
repos_lang = []
for page in repos_list:
    for repo in page:
        repos_lang.append(repo['language'])

len(repos_lang)
```

```python
df_amz = pd.DataFrame({'Repository_Name': repos_name, 'Repository_Language': repos_lang})
df_amz
```
